```python
import pandas as pd
import numpy as np
import statsmodels as sm
import matplotlib.pyplot as plt
import seaborn as sns
from plotnine import *
from plotly import *
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

## preprocess

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DATA340/assignments/Final/data-yyg3l.csv')
data
```

```python
data2 = data.drop(columns='2020 Total Papers', axis=1)
data2['year']='2022'
data2.rename(columns={'2022 Total Papers':'Total Papers'},inplace=True)
data2.head()
```

```python
data3 = data.drop(columns='2022 Total Papers', axis=1)
data3['year']='2020'
data3.rename(columns={'2020 Total Papers':'Total Papers'},inplace=True)
data3.head()
```

```python
combined_data=pd.concat([data2,data3])
```

```python
#combined_2 = combined_data.reset_index()
combined_1 = combined_data.loc[combined_data['Owner Name'] != 'CherryRoad Media']
combined_2 = combined_1.sort_values(by=['Total Papers','year'], ascending=True)
combined_3 = combined_2.tail(20)
combined_3
```

## plot

```python
def generate_colors(color_list, number):
  ''' 
    This function randomly selects n colors from an inputted list
      Input: 
      color_list = list of hexadecimal codes to choose from
      number = number of colors to randomly select from the list
  '''
  s = set()
  while len(s)<number:
    rand_colors = [random.choice(color_list)]
    for x in rand_colors:
        if x not in s: 
          s.add(x)
  return list(s)
```

```python
from plotnine.themes.themeable import panel_background
#specify colors for scatter points
color_discrete_map3={'2022':'rgba(35,71,135,255)','2020':'rgba(238,142,128,255)'}

fig = px.scatter((combined_3), x="Total Papers", y="Owner Name", color='year',width=800, height=600,
                 title='Biggest Media Owners by Number of Newspapers Owned in 2020 and 2022<br><sup> </sup>', 
                 labels={
                     "Owner Name":'OWNER NAME',
                     "Total Papers":'TOTAL PAPERS',
                     "year":" "
                 },
                 color_discrete_map=color_discrete_map3)

for i in combined_3['Owner Name'].unique(): #for each country, connect start and end values
    # filter by country
    df_sub = combined_3[combined_3['Owner Name'] == i]
    fig.add_shape(
        type='line',
        layer='above',
        # connect the two markers
        y0=df_sub['Owner Name'].values[0], x0=df_sub['Total Papers'].values[0],
        y1=df_sub['Owner Name'].values[1], x1=df_sub['Total Papers'].values[1], 
    )

#increase scatter point size
fig.update_traces(marker_size=10)   
fig.update_yaxes(type='category', automargin=True, 
                 range=(-1, len(combined_3['Owner Name'].unique())),
                 ticksuffix = " ",
                 gridcolor='#ebebeb',
                 linecolor='#ebebeb', 
                 title_font_family='Brandon Grotesque',
                 title_font_size=12)
fig.update_xaxes(range=(0,650), 
                 title_font_family='Brandon Grotesque',
                 title_font_size=12,
                 color='black')
fig.update_layout(
   autosize=True,
   yaxis={"dtick":1, "tickfont":dict(family='Courier New')},
   xaxis={"tickfont":dict(family='Courier New', color='black')},
   plot_bgcolor='#FBFBF8',
   paper_bgcolor = '#FBFBF8',
   legend=dict(
    yanchor="top",
    y=1.06,
    xanchor="left",
    x=-0.06,
    orientation="h",
    font_family='Courier New',
    font_color='black'), 
    title_font_family='Gill Sans',
    title_font_color="black"

)
fig.show()
```
